Sampling - Từ liên tục sang rời rạc, bằng lấy sample rate

Nếu sample rate thấp quá thì bị aliasing

```bash
=> f_s >= 2 * f_max
```

How to Anti-aliasing

- This happend when the sr is too low, the data got noised

-> How to solve this

- Oversample the signal

Use Nyquist-Shannon Theorem: where "f_s >= 2 * f_max", Use higer frame rate to helps reduce the aliasing in fast motion
-> This cause many compute

- Smooth the signal (Gaussion Filter / Naive Filter)

Remove detail effects that cause aliasing (Xóa bỏ tần số cao)

-> Cause lose information

Smooth must f_s / 2 with sr larger then twice of the signal

Gaussian IMG Pyramid

- Applying Gaussion Filter to blur the IMG, then reduce it size in even row and column

##Laplacian IMG Pyramid##

for n in i:
    Original - Downsample IMG = Residual

-> How to do

IMG + Gaussian Filter == Blur IMG
IMG - Blur IMG = Residual

Residual + Blur IMG = IMG

````python
import cv2
import matplotlib.pyplot as plt

IMG = r"F:\Class25\INS3155\Sample_IMG.jpg"

Input = cv2.imread(IMG, cv2.IMREAD_COLOR)
blur = cv2.GaussianBlur(Input,(5,5),0)
residual = Input - blur

Recostructed = blur + residual

plt.subplot(131), plt.imshow(Input), plt.title('Original')
plt.subplot(132), plt.imshow(blur), plt.title('Gaussian Blur')
plt.subplot(133), plt.imshow(residual), plt.title('Residual')

plt.show()
````
